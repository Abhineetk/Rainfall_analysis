# Indian Rainfall Analysis

The 2019 Indian floods were a series of floods that affected over thirteen states in late July and early August 2019, due to incessant rains. At least 200 people died and about a million people were displaced. Karnataka and Maharashtra were the most severely affected states.

It was the heaviest monsoon in the last 25 years. More than 1600 people died between June and October 2019.

India being an agriculturally driven economy, it will be interesting to study the rainfall in India in the past decade to give us an idea of the changes in the pattern if there are any.

Let us work on the INDIAN RAINFALL DATA!

## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read the data

```python
rain = pd.read_csv('../data/rainfall in india 1901-2015.csv')
```

## Summary of the data

```python
rain.head()
```

```python
rain.describe()
```

## Inspect the data, find missing values and replace them with appropriate values

### Finding year where all months data is missing

```python
rain.YEAR[rain.iloc[:,2:14].isnull().all(axis=1)]
```

#### there is no such year where all rows are missing

```python
rain.isnull().sum()
```

```python
rain = rain.fillna(rain.mean())
```

```python
rain.isnull().sum()
```

## Plot the mean annual rainfall and note down your observations regarding the same

```python
mean_rainfall =rain.groupby('YEAR')['ANNUAL'].mean()
mean_rainfall.plot(legend = True, label = 'Mean Annual Rainfall', figsize=(10,7))

ma10 = mean_rainfall.rolling(10).mean()
ma10.plot(legend=True,label='Moving Avg. for last 10 years')

plt.xlabel('Year',fontsize=15)
plt.ylabel('Annual Rainfall (in mm)',fontsize=15)
plt.title('Annual Rainfall in India from year 1901-2015', fontsize=20)
plt.grid()
plt.ioff()
plt.show()
```

## Similarly analyze the seasonal rainfall as per subdivisions and note down your observations

```python
seasonal_rain = rain.groupby(['SUBDIVISION',])['Jan-Feb', 'Mar-May','Jun-Sep', 'Oct-Dec'].mean().sort_values('Jun-Sep',ascending=True).plot.bar(width=0.5,edgecolor='k',align='center',stacked=True,figsize=(16,8))
plt.xlabel('rainfall in mm')
plt.grid()
plt.ioff()
plt.show()
```

## With boxplot analyze the distribution of rainfall in various states and onote down your observations

```python
plt.figure(figsize=(15,12))
ax = sns.boxplot(x='ANNUAL', y='SUBDIVISION', data =rain,width=0.8)
ax.tick_params(axis='x',labelsize=20,rotation=0)
ax.tick_params(axis='y',labelsize=20,rotation=0)
plt.grid()
plt.ioff()
plt.show()
```

## Analyze the rainfall through years in Kerala and note down your observations

```python
kerala = rain[rain['SUBDIVISION']=='KERALA']
kerala.groupby('YEAR')['ANNUAL'].mean().plot(ylim=(1000,5000),color='r',marker='o',linestyle='-',linewidth=2,figsize=(12,8))
plt.grid()
plt.ioff()
plt.show()
```

## Read the district wise rainfall data

```python
Dist = pd.read_csv('../data/district wise rainfall normal.csv')
Dist.head(10)
```

## Annual rainfall in different districts of Kerala

```python
kerala_dist = Dist[Dist['STATE_UT_NAME']=='KERALA']

kerala_dist.sort_values('ANNUAL').plot.barh(x='DISTRICT', y='ANNUAL',title="Rainfall in Districts of Kerala",color='orange',
            figsize=(12,8),
           legend=False)
plt.show()
```

## Find out the districts with least rainfall

```python
Dist.sort_values('ANNUAL',ascending=False).tail(10).plot.barh(y='ANNUAL',x='DISTRICT',color='red')
```

```python
(Dist
 .groupby(['STATE_UT_NAME', 'DISTRICT'])
 .ANNUAL.mean()
 .sort_values(ascending=False)
 .tail(10)
 .plot.barh(x='DISTRICT',y='ANNUAL',
           legend=False,
           title='Districts with Minumum Rainfall in India')
)
```

## Similarly, find districts with maximum rainfall

```python
Dist.sort_values('ANNUAL',ascending=True).tail(10).plot.barh(y='ANNUAL',x='DISTRICT',color='green')
```
